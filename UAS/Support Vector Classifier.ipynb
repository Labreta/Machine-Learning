## Data Preprocessing

```python
from sklearn import preprocessing
def label_array(array) :
    encoder = preprocessing.LabelEncoder()
    encoder.fit(array.unique())
    encoded_value = encoder.transform(array)
    return encoded_value
```

```python
X1 = pd.DataFrame({"":label_array(df['Universitas'],)})
X2 = pd.DataFrame({"":label_array(df['Jenis_Program_Studi'],)})
X3 = pd.DataFrame({"":label_array(df['Keluhan_Fisik_Selama_Pembelajaran_Online'],)})
X4 = pd.DataFrame({"":label_array(df['Kendala_Selama_Pembelajaran_Online'],)})
X5 = pd.DataFrame({"":label_array(df['Keluhan_Psikis_Selama_Pembelajaran_Online'],)})

print(X1)
print(X2)
print(X3)
print(X4)
print(X5)
```

```python
y = df['Kepuasan_Terhadap_Pembelajaran_Online']
y
```

```python
pp_df = pd.DataFrame(('Support Vector Classifier', X1, X2, X3, X4, X5)).transpose()

pp_df.columns = ['Model', 'Universitas', 'Jenis Program Studi', 'Keluhan Fisik Selama Pembelajaran Online', 'Kendala Selama Pembelajaran Online', 'Keluhan Psikis Selama Pembelajaran Online']
pp_df
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(pp_df, y, test_size=0.2, random_state=100)
```

# Modeling 

### Support Vector Classifier

```python
from sklearn.svm import SVC
import numpy as np
svc=SVC(C=100.0) 
svc.fit(X_train, y_train)
```

Training the model

```python
y_svc_train_pred = svc.predict(X_train)
y_svc_test_pred = svc.predict(X_test)
```

```python
y_svc_train_pred
```

```python
y_svc_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

svc_train_mse = mean_squared_error(y_train, y_svc_train_pred)
svc_train_r2 = r2_score(y_train, y_svc_train_pred)

svc_test_mse = mean_squared_error(y_test, y_svc_test_pred)
svc_test_r2 = r2_score(y_test, y_svc_test_pred)
```

```python
print ('RF MSE (Train)', svc_train_mse)
print ('RF r2 (Train)', svc_train_r2)
print ('RF MSE (Test)', svc_test_mse)
print ('RF r2 (Test)', svc_test_r2)
```

```python
svc_result = pd.DataFrame(('SVM', svc_train_mse, svc_train_r2, svc_test_mse, svc_test_r2)).transpose()
svc_result.columns = ['Model', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
svc_result
```
