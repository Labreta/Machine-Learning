## Data Preprocessing

```python
from sklearn import preprocessing
def label_array(array) :
    encoder = preprocessing.LabelEncoder()
    encoder.fit(array.unique())
    encoded_value = encoder.transform(array)
    return encoded_value
```

```python
X1 = pd.DataFrame({"":label_array(df['Universitas'],)})
X2 = pd.DataFrame({"":label_array(df['Jenis Program Studi'],)})
X3 = pd.DataFrame({"":label_array(df['Keluhan Fisik Selama Pembelajaran Online'],)})
X4 = pd.DataFrame({"":label_array(df['Kendala Selama Pembelajaran Online'],)})
X5 = pd.DataFrame({"":label_array(df['Keluhan Psikis Selama Pembelajaran Online'],)})

print(X1.to_string(index=False, header=False))
print("\n")
print(X2.to_string(index=False, header=False))
print("\n")
print(X3.to_string(index=False, header=False))
print("\n")
print(X4.to_string(index=False, header=False))
print("\n")
print(X5.to_string(index=False, header=False))
```

```python
y = df['Kepuasan Terhadap Pembelajaran Online']
y
```

```python
pp_df = pd.concat([X1, X2, X3, X4, X5], axis=1)
pp_df.columns = ['Universitas', 'Jenis Program Studi', 'Keluhan Fisik Selama Pembelajaran Online', 'Kendala Selama Pembelajaran Online', 'Keluhan Psikis Selama Pembelajaran Online']
pp_df.insert(0, "Model", "Support Vector Classifier")
pp_df = pp_df.drop('Model', axis=1)
pp_df
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(pp_df, y, test_size=0.2, random_state=100)
```

```python
X_test
y_train
```

# Modeling 

### Support Vector Classifier

```python
from sklearn.svm import SVC
import numpy as np
svc=SVC(C=100.0) 
svc.fit(X_train, y_train)
```

## Training the model

```python
y_svc_train_pred = svc.predict(X_train)
y_svc_test_pred = svc.predict(X_test)
```

```python
y_svc_train_pred
```

```python
y_svc_test_pred
```

## Evaluasi Model

```python
from sklearn.metrics import accuracy_score, precision_score

y_svc_train_pred = svc.predict(X_train)
y_svc_test_pred = svc.predict(X_test)

svc_train_accuracy = accuracy_score(y_train, y_svc_train_pred)
svc_test_accuracy = accuracy_score(y_test, y_svc_test_pred)

svc_train_precision = precision_score(y_train, y_svc_train_pred, average='weighted')
svc_test_precision = precision_score(y_test, y_svc_test_pred, average='weighted')

print('Train Accuracy:', svc_train_accuracy)
print('Test Accuracy:', svc_test_accuracy)
print('Train Precision:', svc_train_precision)
print('Test Precision:', svc_test_precision)
```

```python
svc_result = pd.DataFrame(('SVM', svc_train_accuracy, svc_test_accuracy, svc_train_precision, svc_test_precision)).transpose()
svc_result.columns = ['Model', 'Train Accuracy', 'Test Accuracy', 'Test Precision', 'Test Precision']
svc_result
```
