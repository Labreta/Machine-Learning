## Data Preprocessing

```python
from sklearn import preprocessing
def label_array(array) :
    encoder = preprocessing.LabelEncoder()
    encoder.fit(array.unique())
    encoded_value = encoder.transform(array)
    return encoded_value
```

```python
X1 = pd.DataFrame({"":label_array(df['Universitas'],)})
X2 = pd.DataFrame({"":label_array(df['Jenis_Program_Studi'],)})
X3 = pd.DataFrame({"":label_array(df['Keluhan_Fisik'],)})
X4 = pd.DataFrame({"":label_array(df['Kendala_Online'],)})
X5 = pd.DataFrame({"":label_array(df['Keluhan_Psikis'],)})

print("")
print(X1.to_string(index=False, header=False))

print("")
print(X2.to_string(index=False, header=False))

print("")
print(X3.to_string(index=False, header=False))

print("")
print(X4.to_string(index=False, header=False))

print("")
print(X5.to_string(index=False, header=False))
```

```python
y = df['Kepuasan_Online']
print(y.to_string(index=False, header=False))
```

```python
pp_df = pd.concat([X1, X2, X3, X4, X5], axis=1)
pp_df.columns = ['Universitas', 'Jenis_Program_Studi', 'Keluhan_Fisik', 'Kendala_Online', 'Keluhan_Psikis']
pp_df.insert(0, "Model", "Random Forest Classifier")
pp_df = pp_df.drop('Model', axis=1)
pp_df
```

```python
new_data = pd.read_csv('data_training4.csv')
```

```python
new_df1 = df['Kualitas_Dosen'].values
new_df2 = df['Kepuasan_Infrastruktur'].values
new_df3 = df['Performa_Akademik'].values
```

```python
new_df = pd.DataFrame({
    'Kualitas_Dosen': new_df1 , 
    'Tingkat_Infrastruktur': new_df2,
    'Performa_Akademik': new_df3 
})

new_df
```

```python
df = pd.concat([pp_df, new_df], axis=1)
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.5, random_state=100)
```

```python
X_train
```

```python
X_test
y_train
```

# Modeling 

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

## Training the model

```python
y_rfc_train_pred = rfc.predict(X_train)
y_rfc_test_pred = rfc.predict(X_test)
```

```python
y_rfc_train_pred
```

```python
y_rfc_test_pred
```

## Evaluasi Model

```python
#Input Model
from sklearn.metrics import accuracy_score, precision_score

rfc_train_accuracy = accuracy_score(y_train, y_rfc_train_pred)
rfc_test_accuracy = accuracy_score(y_test, y_rfc_test_pred)

rfc_train_precision = precision_score(y_train, y_rfc_train_pred, average='weighted', zero_division=0)
rfc_test_precision = precision_score(y_test, y_rfc_test_pred, average='weighted', zero_division=0)

print('Train Accuracy:', rfc_train_accuracy)
print('Test Accuracy:', rfc_test_accuracy)
print('Train Precision:', rfc_train_precision)
print('Test Precision:', rfc_test_precision)
```

```python
rfc_result = pd.DataFrame(('Random Forest Classifier', rfc_train_accuracy, rfc_test_accuracy, rfc_train_precision, rfc_test_precision)).transpose()
rfc_result.columns = ['Model', 'Accuracy Train', 'Accuracy Test', 'Precision Train', 'Precision Test']
rfc_result
```
