## Data Preparation

```python
y = df['Kepuasan_Terhadap_Pembelajaran_Online']
y
```

```python
X = df.drop('Kepuasan_Terhadap_Pembelajaran_Online', axis=1)
X
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=100)
```

```python
X_train
```

```python
X_test
```

## Modelling

### Random Forest Classifier

Training The Model 

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

Making The Prediction To The Model

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
y_rf_train_pred
```

```python
y_rf_test_pred
```

### Evaluate The Model 

```python
y_train
```

```python
y_rf_train_pred
```

### Evaluate The Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print ('RF MSE (Train)', rf_train_mse)
print ('RF r2 (Train)', rf_train_r2)
print ('RF MSE (Test)', rf_test_mse)
print ('RF r2 (Test)', rf_test_r2)
```

```python
rf_result = pd.DataFrame(('Random Forest Classifier', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2)).transpose()
rf_result.columns = ['Model', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
rf_result
```
